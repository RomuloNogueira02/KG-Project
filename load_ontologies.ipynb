# Load Ontology and extract labels

```python
from owlready2 import *
import os
```

```python
# Constants
ONTOLOGIES = os.getcwd() + "/ontologies/"
```

```python
# Functions 
def load_ontology(folder, name):
    return get_ontology(ONTOLOGIES + folder + "/" + name).load()

def get_labels(ontology):
    # Get labels from ontology that are not empty
    raw_labels = list(map(lambda entity: entity.label[0] if len(entity.label) > 0 else "", ontology.classes()))
    filtered_labels = list(filter(lambda label: label != "", raw_labels))
    return filtered_labels
```

```python
ontology1 = load_ontology("anatomy", "human.owl")
ontology2 = load_ontology("anatomy", "mouse.owl")
```

```python
print("Classes in Human Ontology:", len(list(ontology1.classes())))
print("Classes in Mouse Ontology:", len(list(ontology2.classes())))
```

```python
print("Labels in Human Ontology:", get_labels(ontology1))
print("Labels in Mouse Ontology:", get_labels(ontology2))
```
