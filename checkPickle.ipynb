```python
import pickle
import pandas as pd
```

```python
with open('alignment_result.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
print(len(data))
```

```python
final_result = {}
for tuplo, score in data.items():
    if type(score) == tuple:
        avg_score = sum(score) / len(score)
        if avg_score >= 0.7:
            final_result[tuplo] = avg_score
    else:
        final_result[tuplo] = score
```

```python
len(final_result)
```

```python
final_result
```

```python
triplos = [(key[0], key[1], value) for key, value in final_result.items()]
```

```python
df = pd.DataFrame(triplos, columns=['ontology 1', 'ontology 2', 'score'])
df
```

```python
toDrop = set()
for index, row in df.iterrows():
    # print(row['ontology 1'], row['ontology 2'], row['score'])
    duplicates = df[df["ontology 1"] == row['ontology 1']]

    if len(duplicates) > 1:
        print(duplicates)
        # print(duplicates[duplicates["score"] < max(duplicates["score"])].index.values.tolist())
        indexes = set(duplicates[duplicates["score"] < max(duplicates["score"])].index.values.tolist())
        print(indexes)
        toDrop |= indexes
        print("---")
len(toDrop)
```

```python
droped_left = df.drop(toDrop)
droped_left
```

```python
toDrop2 = set()
for index, row in droped_left.iterrows():
    duplicates = droped_left[droped_left["ontology 2"] == row['ontology 2']]

    if len(duplicates) > 1:
        print(duplicates)
        # print(duplicates[duplicates["score"] < max(duplicates["score"])].index.values.tolist())
        indexes = set(duplicates[duplicates["score"] < max(duplicates["score"])].index.values.tolist())
        print(indexes)
        toDrop2 |= indexes
        print("---")
len(toDrop2)
```

```python
droped_left_right = droped_left.drop(toDrop2)
droped_left_right.sort_values(by='score', ascending=False)
```

```python
droped_left_right.to_excel('alignment_result.xlsx', index=False)
```
