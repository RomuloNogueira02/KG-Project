```python
import pickle
```

```python
with open('alignment_result.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
print(len(data))
```

```python
final_result = {}
for tuplo, score in data.items():
    if type(score) == tuple:
        avg_score = sum(score) / len(score)
        if avg_score >= 0.7:
            final_result[tuplo] = avg_score
    else:
        final_result[tuplo] = score
```
